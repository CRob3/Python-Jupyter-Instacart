### Create new column sorted by prices
<a id='price_column'></a>

```python
# Create the function
def price_label(row):
    if row ['prices'] <= 5:
        return 'Low-range product'
    elif (row ['prices'] > 5) and (row ['prices'] <= 15):
        return 'Mid-range product'
    elif row ['prices'] > 15:
        return 'High-range product'
    else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis = 1)
```

### For Loops
<a id='for_loops'></a>

```python
ords_prds_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
result = []
for value in ords_prds_merged['orders_day_of_week']:
    if value == 0:
        result.append("Busiest Day")
    elif value == 4:
        result.append ("Least Busy")
    else:
        result.append("Regularly Busy")
```

```python
ords_prds_merged['busiest_day'] = result
```

```python
ords_prds_merged.head()
```

```python
ords_prds_merged['busiest_day'].value_counts(dropna = False)
```

```python
# Create two busiest days label
busy = []
for value in ords_prds_merged['orders_day_of_week']:
    if value == 0 or value == 1:
        busy.append("Busiest Days")
    elif value == 4 or value == 3:
        busy.append("Slowest Days")
    else:
        busy.append("Regular Day")
```

```python
busy
```

```python
ords_prds_merged['shopping_days'] = busy
```

### Create an hours category column
<a id='hours_category'></a>

```python
ords_prds_merged['order_hour_of_day'].value_counts(dropna = False)
```

#### I will divide the hours into most (7 digit count), average (6 digit count) and fewest (5 digit count) number of orders.

```python
time = []
for value in ords_prds_merged['order_hour_of_day']:
    if value in {10,11,14,15,13,12,16,9,17,8,18,19}:
        time.append("Most orders")
    elif value in {20,7,21,22,23,6,0,1}:
        time.append("Average orders")
    elif value in {5,2,4,3}:
        time.append("Fewest orders")
```

```python
time
```

```python
ords_prds_merged['busy_hours'] = time
```

```python
ords_prds_merged.head(10)
```

```python
ords_prds_merged['busy_hours'].value_counts(dropna = False)
```
