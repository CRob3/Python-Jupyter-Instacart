### Find aggregated mean of order_number by department_id
<a id='mean_order_dept'></a>

```python
ords_prds_merge.groupby('department_id').agg({'order_number':['mean']})
```

#### Analyzing the entire dataframe gives the results for all of the departments.  17 (household goods) now has a higher average (15.6 instead of 11) and 4 (produce) now has a lower mean (17.8 instead of 18.8).  Clearly including all of the results alters the statistics.

### Loyalty flag - was already created when following the exercises in the lesson with the following code:
<a id='loyalty'></a>

```python
ords_prds_merge['max_order'] = ords_prds_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prds_merge.loc[ords_prds_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal Customer'
```

```python
ords_prds_merge.loc[(ords_prds_merge['max_order'] <= 40) & (ords_prds_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular Customer'
```

```python
ords_prds_merge.loc[ords_prds_merge['max_order'] <= 10, 'loyalty_flag'] = 'New Customer'
```

```python
ords_prds_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Check basic stats of product prices for the three loyalty categories
ords_prds_merge.groupby('loyalty_flag').agg({'prices':['mean', 'min', 'max']})
```

### Group users by average price of all their orders
<a id='group_price'></a>

```python
ords_prds_merge['avg_spent'] = ords_prds_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prds_merge['avg_spent'].head(30)
```

```python
ords_prds_merge['avg_spent'].describe()
```

```python
# Create flags based on avg_spent (< 10 = low spender, >= 10 = high spender)
ords_prds_merge.loc[ords_prds_merge['avg_spent'] < 10, 'spenders'] = 'Low Spender'
```

```python
ords_prds_merge.loc[ords_prds_merge['avg_spent'] >= 10, 'spenders'] = 'High Spender'
```

```python
ords_prds_merge[['user_id', 'avg_spent', 'spenders']].head(30)
```

```python
ords_prds_merge['spenders'].value_counts(dropna = False)
```

```python
ords_prds_merge.shape
```

### Create a frequency flag based on days_since_prior_order median
<a id ='frequency'></a>

```python
ords_prds_merge['med_frequency'] = ords_prds_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prds_merge['med_frequency'].head(30)
```

```python
# Create frequency flag: > 20 = non-frequent customer, >10 and <=20 = regular customer, <=10 = Frequent customer
ords_prds_merge.loc[ords_prds_merge['med_frequency'] > 20, 'frequent'] = 'Non-Frequent Customer'
```

```python
ords_prds_merge.loc[(ords_prds_merge['med_frequency'] > 10) & (ords_prds_merge['med_frequency'] <=20), 'frequent'] = 'Regular Customer'
```

```python
ords_prds_merge.loc[ords_prds_merge['med_frequency'] <= 10, 'frequent'] = 'Frequent Customer'
```

```python
ords_prds_merge[['user_id', 'med_frequency', 'frequent']].head(30)
```

```python
ords_prds_merge['frequent'].value_counts(dropna = False)
```
